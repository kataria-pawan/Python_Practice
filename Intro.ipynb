```python
# right triangle pattern
for i in range(1,6):
    for j in range(i):
        print("*", end = " ")
    print()
```

```python
for i in range(1,6):
    for j in range(i):
        print(j, end=" ")
    print()
```

```python
# inverted right triangle
for i in range(6,0,-1):
    for j in range(i):
        print("*", end= " ")
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i):
        print(j, end= " ")
    print()
```

```python
# pyramid pattern
for i in range(1,6):
    for j in range(6-i):
        print(" ", end = " ")
    for k in range(1,2*i):
        print("*", end= " ")
    print()
```

```python
for i in range(6,0,-1):
    for j in range(6-i):
        print(" ", end = ' ')
    for k in range(1,2*i):
        print("*", end= " ")
    print()
```

```python
# diamond pattern
for i in range(0,6):
    for j in range(6-i):
        print(" ", end= " ")
    for k in range(1, 2*i):
        print("*", end = ' ')
    print()
for i in range(6,0,-1):
    for j in range(6-i):
        print(" ", end= " ")
    for k in range(1, 2*i):
        print("*", end= " ")
    print()
```
